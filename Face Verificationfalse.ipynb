```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread('color.jpg')
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

gray = rgb2gray(img)
plt.imshow(gray, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
import numpy as np
from PIL import Image
from skimage import color

def image_to_vector(image_path):
    # Open the image file
    image = Image.open(image_path)
    image = image.convert('L')
    image = image.resize((500, 500))
    image_vector = np.asarray(image).flatten()
    image_vector = image_vector / np.linalg.norm(image_vector)
    return image_vector

def cosine_similarity(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    similarity = dot_product / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
    return similarity
```

```python
vector1 = image_to_vector('/content/Cristiano_Ronaldo_2018.jpg')
vector2 = image_to_vector('/content/gettyimages-971463110.jpg')

similarity_score = cosine_similarity(vector1, vector2)
print(f'Cosine Similarity: {similarity_score}')
if (similarity_score>0.8):
  print('image is real')
```
