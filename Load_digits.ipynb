```python
from sklearn.datasets import load_digits
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
digits = load_digits()
df=pd.DataFrame(digits.data,columns=digits.feature_names)
df.head()
```

```python
x=df.drop("target",axis="columns")
y=df.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scaler=MinMaxScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_predicted = model.predict(x_test)
model.score(x_test,y_test)
accuracy=accuracy_score(y_test,y_predicted)
accuracy
```

```python
model2=tree.DecisionTreeClassifier()
model2.fit(x_train,y_train)
```

```python
y_predicted = model2.predict(x_test)
model2.score(x_test,y_test)
accuracy2=accuracy_score(y_test,y_predicted)
round(accuracy2*100,2)
```

```python
confusion_matrix(y_test,y_predicted)
```

```python
model3 = RandomForestClassifier()
model3.fit(x_train, y_train)
```

```python
y_predicted = model3.predict(x_test)
accuracy3 = accuracy_score(y_test, y_predicted)
print("Random Forest Accuracy: %",round(accuracy3*100,2))
```

```python
model4 = GradientBoostingClassifier()
model4.fit(x_train, y_train)
```

```python
y_predicted = model4.predict(x_test)
accuracy4 = accuracy_score(y_test, y_predicted)
print(" Boosting Accruracy: %",round(accuracy4*100,2))
```
