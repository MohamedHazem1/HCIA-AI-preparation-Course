```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
######################################
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
```

```python
filename='/content/housing.csv'
column_names = ['A', 'B', 'C', 'ZZZZ', 'D', 'E', 'F ','G','H','I','J','K','L','Price']
df=pd.read_csv(filename,names=column_names,sep=" ",skipinitialspace=True)
df.drop(['ZZZZ'],axis='columns',inplace=True)
df.head()
```

```python
mms=MinMaxScaler()
mms.fit(df)
df2=mms.transform(df)
```

```python
df3=pd.DataFrame(df2)
df3.head()
```

```python
Y =df3.iloc[:,12].values
X=df3.iloc[:,:11].values
X
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
modeln =LinearRegression()
modeln.fit(x_train,y_train)
y_pred=modeln.predict(x_test)
print("Linear Regression")
print("Train Score")
modeln.score(x_train, y_train)
```

```python
print("Test Score")
modeln.score(x_test, y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#specify degree of 3 for polynomial regression model
#include bias=False means don't force y-intercept to equal zero
poly = PolynomialFeatures(degree=2, include_bias=False)

#reshape data to work properly with sklearn
poly_features = poly.fit_transform(X)

#fit polynomial regression model
poly_reg_model = LinearRegression()
poly_reg_model.fit(poly_features, Y)
```

```python
mse1 = mean_squared_error(Y, poly_reg_model.predict(poly_features))
r21 = r2_score(Y, poly_reg_model.predict(poly_features))
print("MSE: ", mse1)
print("R2: ", r21)
```

```python
model = Sequential()
n=11 #No. of inputs
model.add(Dense(128, activation='relu', input_shape=(n,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='relu')) # Last Stage : should be the number of the output

model.compile(optimizer='sgd',loss='mse',metrics=['mse'])
```

```python
trained=model.fit(x_train,y_train,epochs=200,batch_size=32,validation_split=0.2,verbose=1)
```
