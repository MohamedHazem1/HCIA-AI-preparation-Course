```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/salary_data.csv')
df
```

```python
X= df.iloc[:,0]
y=df.iloc[:,1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
def model (X,A,B):
  return A*X+B
```

```python
def mse(y, y_hat):
  return 1/(2*len(y)) * np.sum((y_hat - y)**2)
```

```python
def gd(A,B,X,y,lr=0.01):
  A=A-(lr/len(y))*np.dot(np.array((model(X,A,B)-y)).T,X)
  B=B-(lr/len(y))*np.array(model(X,A,B)-y)
  return A,B
```

```python
def epoch(X,y,A,B,lr):
  A, B= gd(A,B,X,y,lr=lr)
  loss=mse(y,model(X,A,B))
  return A,B , loss
```

```python
X = np.array(X).reshape(len(X),1)
y = np.array(y).reshape(len(y),1)
A = np.random.randn(1,1)
B = np.random.randn(1,1)
lr = 0.01
epochs=100
losses=[]
for i in range(epochs):
  A,B,loss=epoch(X,y,A,B,lr)
  losses.append(loss)

plt.plot(np.arange(100),losses)
```
